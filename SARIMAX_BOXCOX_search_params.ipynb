```python
# -*- coding: utf-8 -*-
from pandas import read_csv, DataFrame
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.tsa.arima_model import ARIMA
from statsmodels.tsa.seasonal import seasonal_decompose
from itertools import product
import warnings
from sklearn.metrics import r2_score

import pandas as pd
import numpy as np
from scipy import stats

import statsmodels.api as sm

def invboxcox(x, lmbda = 0):
    if lmbda == 0:
        return(np.exp(x))
    else:
        return(np.exp(np.log(lmbda * x + 1) / lmbda))
```

```python
dataset = read_csv('training.csv',
                     ',',
                     index_col=['Date'],
                     parse_dates=['Date'],
                     dayfirst=True)
```

```python
dataset['Value_boxcox'], lmbda = stats.boxcox(dataset.Value)
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

Qs = range(0, 3)
qs = range(0, 5)
Ps = range(0, 5)
ps = range(0, 6)
d = 1
D = 0
S = range(12, 13)

parameters = product(ps, qs, Ps, Qs, S)
parameters_list = list(parameters)

results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')
i = 0

for param in parameters_list:
    i += 1
    try:
        model = SARIMAX(dataset.Value_boxcox, 
                        order=(param[0], d, param[1]), 
                        seasonal_order = (param[2], 
                                          D, 
                                          param[3],
                                          param[4])).fit(disp = -1)
    except:
        print 'waring parametrs:', param
        continue
    print 'Computing...' + str(i)
    if model.aic < best_aic:
        best_model = model
        best_aic = model.aic
        best_param = param
        print param, model.aic
    results.append([param, model.aic])
    

warnings.filterwarnings('default')

result_table = DataFrame(results)
result_table.columns = ['parameters', 'aic']
print result_table.sort_values(by = 'aic', ascending = [True]).head()
```
